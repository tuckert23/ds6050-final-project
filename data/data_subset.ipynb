# Code to Subset the Data for Upload

```python
import pandas as pd
```

```python
data = pd.read_csv("./NFL_Play_by_Play_2009-2018_(v5).csv")
data.head()
```

```python
data["game_year"] = data["game_date"].apply(lambda x: x[:4]).astype(int)
data.head()
```

```python
data.shape
```

```python
data_subset = data[data["game_year"] > 2017]
data_subset.shape
```

```python
data_subset.to_csv("./data_subset.csv")
```
